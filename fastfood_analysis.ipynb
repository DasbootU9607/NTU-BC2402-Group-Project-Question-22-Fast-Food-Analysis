```python
# fix_notebook_issue.py
import os
import shutil

def fix_notebook():
    print("=== Fixing Notebook File ===")
    
    # 检查当前 fastfood_analysis.ipynb 的大小
    current_size = os.path.getsize("fastfood_analysis.ipynb")
    print(f"Current fastfood_analysis.ipynb size: {current_size} bytes")
    
    # 检查是否有备份或原始文件在其他位置
    search_paths = [
        os.getcwd(),
        os.path.expanduser("~"),
        os.path.expanduser("~/Desktop"), 
        os.path.expanduser("~/Documents"),
        os.path.expanduser("~/Downloads")
    ]
    
    print("\n=== Searching for the real notebook file ===")
    real_notebook_path = None
    
    for path in search_paths:
        if os.path.exists(path):
            for root, dirs, files in os.walk(path):
                for file in files:
                    if file == "fastfood_analysis.ipynb":
                        full_path = os.path.join(root, file)
                        size = os.path.getsize(full_path)
                        print(f"Found: {full_path} - {size} bytes")
                        if size > 100000:  # 如果文件大于 100KB，很可能是真的
                            real_notebook_path = full_path
                            print(f"✅ Likely real notebook: {size} bytes")
    
    if real_notebook_path and os.path.getsize(real_notebook_path) > current_size:
        # 替换文件
        shutil.copy2(real_notebook_path, "fastfood_analysis.ipynb")
        print(f"✅ Replaced with real notebook from: {real_notebook_path}")
    else:
        print("No larger notebook file found. The current file might be correct.")

# 运行修复
fix_notebook()

# 验证修复结果
print("\n=== Verification ===")
final_size = os.path.getsize("fastfood_analysis.ipynb")
print(f"Final fastfood_analysis.ipynb size: {final_size} bytes")
```

```python
# copy_correct_notebook.py
import os
import shutil
import glob

def find_and_copy_real_notebook():
    print("=== Searching for the REAL 4,476 KB notebook file ===")
    
    # 在常见位置搜索大的 .ipynb 文件
    search_locations = [
        os.path.expanduser("~"),  # 用户主目录
        os.path.expanduser("~/Desktop"),
        os.path.expanduser("~/Documents"), 
        os.path.expanduser("~/Downloads"),
        "C:/",  # 整个 C 盘搜索（可能需要时间）
    ]
    
    large_notebooks = []
    
    for location in search_locations:
        if os.path.exists(location):
            print(f"Searching in: {location}")
            try:
                # 搜索所有 .ipynb 文件
                for root, dirs, files in os.walk(location):
                    for file in files:
                        if file.endswith('.ipynb'):
                            full_path = os.path.join(root, file)
                            try:
                                size = os.path.getsize(full_path)
                                if size > 4000000:  # 大于 4MB 的文件
                                    large_notebooks.append((full_path, size))
                                    print(f"🎯 FOUND LARGE NOTEBOOK: {file} - {size} bytes")
                            except:
                                continue
                    # 限制搜索深度，避免耗时太长
                    if len(large_notebooks) > 10:
                        break
            except:
                continue
    
    if large_notebooks:
        print(f"\n=== Found {len(large_notebooks)} large notebook files ===")
        for i, (path, size) in enumerate(large_notebooks, 1):
            print(f"{i}. {path} - {size} bytes")
        
        # 复制最大的文件
        largest = max(large_notebooks, key=lambda x: x[1])
        source_path, size = largest
        
        print(f"\n📋 Copying largest notebook: {os.path.basename(source_path)}")
        shutil.copy2(source_path, "fastfood_analysis.ipynb")
        print(f"✅ Copied {size} bytes to fastfood_analysis.ipynb")
        
        # 验证
        new_size = os.path.getsize("fastfood_analysis.ipynb")
        print(f"✅ New file size: {new_size} bytes")
    else:
        print("❌ No large notebook files found")
        print("\n请手动查找并复制正确的文件：")
        print("1. 找到你原始的分析 notebook 文件")
        print("2. 复制到: C:\\Users\\CHMEI\\ntu_bc2402_project\\fastfood_analysis.ipynb")

# 运行搜索和复制
find_and_copy_real_notebook()
```

```python
# check_recent_notebooks.py
import os
import glob

def find_recent_notebooks():
    print("=== 查找最近使用的 notebook 文件 ===")
    
    # 检查当前用户目录下的所有 .ipynb 文件
    user_dir = os.path.expanduser("~")
    print(f"搜索目录: {user_dir}")
    
    # 使用 glob 快速查找
    notebook_files = []
    for file_path in glob.glob(os.path.join(user_dir, "**", "*.ipynb"), recursive=True):
        try:
            size = os.path.getsize(file_path)
            notebook_files.append((file_path, size))
        except:
            continue
    
    # 按文件大小排序
    notebook_files.sort(key=lambda x: x[1], reverse=True)
    
    print(f"\n找到 {len(notebook_files)} 个 notebook 文件:")
    for i, (path, size) in enumerate(notebook_files[:10]):  # 显示前10个最大的
        file_name = os.path.basename(path)
        print(f"{i+1}. {file_name} - {size} bytes")
        print(f"   位置: {path}")
    
    return notebook_files

# 运行搜索
notebooks = find_recent_notebooks()

if notebooks:
    print(f"\n=== 建议 ===")
    largest_file = notebooks[0]
    print(f"最大的文件: {os.path.basename(largest_file[0])} - {largest_file[1]} bytes")
    
    if largest_file[1] > 100000:  # 如果大于100KB
        print("这可能是你要找的分析文件！")
    else:
        print("所有文件都很小，可能需要在其他位置查找")
```
